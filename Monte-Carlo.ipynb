```python
import random
import numpy as np
import matplotlib.pyplot as plt
random.seed(10)
g = 1.5
R = [0.70,0.69,0.685,0.68,0.675,0.67,0.665,0.655,0.65,0.64,0.62]

periods = [i+1 for i in range(10000)]
worths = []
for i in R:
    lis = [g,i]
    net_worth = 1
    worth_t = []
    for j in periods:
        net_worth = net_worth * random.choice(lis)
        worth_t.append(net_worth)
    worths.append(worth_t)
```

```python
plt.style.use('classic')
%matplotlib inline
import numpy as np
ax = plt.axes(yscale='log')
ax.grid()
for i in worths:
    plt.plot(periods,i)
plt.legend(R,bbox_to_anchor=(1.01, 1), loc='upper left', borderaxespad=0)
plt.ylabel("Net Worth",size=14)
plt.xlabel("Periods n",size=14)
plt.ylim(bottom = 10**-110)
plt.title("pg2738 - Priyesh")
plt.gcf().set_size_inches(12,6)
plt.savefig("HW1 Que1 PG2738-2.png")
plt.show()
```
